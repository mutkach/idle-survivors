```python
import gymnasium_env 
import gymnasium as gym
import numpy as np
import skimage
import matplotlib.pyplot as plt
import gymnasium_env.envs


env = gymnasium_env.envs.VampireWorldEnv(render_mode=None, size=5)
env.reset()
```

```python
rewards = []
for i in range(512):
    row = []
    for j in range(512):
        row.append(env.reward_for_place(np.array([i,j])))
    rewards.append(row)
```

```python
plt.imshow(np.array(rewards),  norm='symlog')
```
