```python
import idle_vampires
import numpy as np
import matplotlib.pyplot as plt
import skimage
from itertools import  product
import idle_vampires 
import gymnasium as gym
```

```python
env = gym.make(
  "idle_vampires/VampireWorld-v0", render_mode=None, window_size=512, num_enemies=5, movement="stick"
)
```

```python
obs, info = env.reset()


x,y = obs['agent'].astype(int)
en=obs['enemies'] 
print(en, (x,y))

#a = np.zeros((512,512,3)).astype(int)
wdth = 100
step = wdth//2
# colors = [
#     [255,255,0],
#     [0,255,255],
#     [255,0,255],
#     [0,255,255],
#     [0,0,255],
#     [0,255,0],
#     [255,0,0],
#     [0,122,122],
#     [122,0,122],
# ]
counts = []
for i,(ox,oy) in enumerate(product([-1,0,1], [-1,0,1])):
    #x+ox*wdth-step:x+ox*wdth+step, y+oy*wdth-step:y+oy*wdth+step) 
    if ox == 0 and oy == 0:
        continue
    h=np.linalg.norm(en-np.array([x+ox*wdth, y+oy*wdth]), ord=1, axis=1)
    print("H:", h)
    #print("DIFF:", np.array([x+ox*wdth, y+oy*wdth] - en))
    counts.append((h<wdth).sum())






print(counts)
```
